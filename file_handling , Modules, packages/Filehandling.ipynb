```python
#purpose - store and retrieve data or information

'''
data - collection of information
types of data
          1. structured data - data that is organized in a specific format - table format - xlsx, csv, sql, etc
            2. unstructured data - data that is not organized in a specific format - media files, images, videos, etc, facebook posts and comments, doc, pdf, ppt   etc
            3. semi-structured data - data that is partially organized in a specific format - json, xml, html
File - collection of data or information stored in a specific format, we can fetch and process the data.

types of files
           1. Text Files  
            - web standards - html,xml,css,js
            - source code - c,app,js,py,java, etc
            - document - txt,tex,rtf etc
            - tabular data - csv, tsv, etc
            - configuration - ini, json, yaml, etc
           2. Binary Files
            - images - jpg, png, gif, etc
            - audio - mp3, wav, etc
            - video - mp4, avi, etc
            - executable - exe, bin, etc
            - archive - zip, tar, etc
            - database - db, sql, etc
            -document - pdf, docx, pptx, etc

file handling - process the data in a file
file handling operations
1. create - create a file
2. open - open a file
3. read - read data from a file
4. write - write data to a file
5. append - add data to a file
6. close - close a file
7. delete - delete a file
8. rename - rename a file
9. copy - copy a file

'''
```

```python
# open() - open a file
 #file_instance=open(file_name,mode)

#hm_txt=open(absolute_path / relative_path,file_mode)
hm_txt=open('C:/python/pythonBasics/fib.py','w') # absolute path
hm_txt=open('fib.py','w') # relative path


# file modes
'''
r,w,a - read, write, append
r+ - read, write, w+ - write and read, a+ - append, read read and write

rb,wb,ab - read, write, append in binary mode
rb+,wb+,ab+ - read and write in binary mode

'''
```

```python
hm_txt=open('test.txt','w+')
hm_txt.write('hello world')
print(hm_txt.tell()) # returns the current position of the file pointer
hm_txt.seek(0) # move the file pointer to the beginning of the file
read_result=hm_txt.read()
print(read_result)
hm_txt.close()

# seek(0) - move the file pointer to the beginning of the file
# seek(1) - move the file pointer to the second byte of the file
# seek(2) - move the file pointer to the third byte of the file
# tell() - returns the current position of the file pointer
```

```python
# read the files
hm_txt=open('fib.py','r')
print(hm_txt.readline())

hm_txt.close()
```

```python
# read the files
hm_txt=open('fib.py','r')
print(hm_txt.readlines())

hm_txt.close()
```

```python
hm_txt=open('test.txt','a')
hm_txt.write('hello world')
hm_txt.writelines(['hello world\n','hello world\n','hello world\n'])
hm_txt.close()
```

```python
# open() the file with keyword with statement - best practice to handle files

with open('test.txt','r') as hm_txt:
    print(hm_txt.read())
    print(hm_txt.tell())
    hm_txt.seek(0)
    print(hm_txt.readlines())
```
